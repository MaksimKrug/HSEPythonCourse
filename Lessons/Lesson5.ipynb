### DataFrame: Merging

```python
# Create a sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': ['foo', np.nan, np.nan, 'bar', 'foo'],
    'C': [0.1, 0.2, 0.3, np.nan, 0.5],
    'D': [True, False, True, False, True]
}

df = pd.DataFrame(data)
df
```

```python
display(df.dropna(axis=0, inplace=False))
display(df.dropna(axis=1, inplace=False))
```

```python
df2 = pd.DataFrame({"A": [4,5,6,7,8], "B": ["new","new","new","new","new"]})
df2
```

```python
df
```

```python
pd.concat((df, df2), axis=0, ignore_index=True)
```

```python
df.join(df2, on="A", how="left", rsuffix="_r")
```

```python
df.merge(df2, on="A", how="left", indicator=True)
```
